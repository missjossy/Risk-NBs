# Get From SF

```python
df=pd.read_csv('Uganda - Marketing_Growth 360 Report - 2025 - Overview May 2025 (1).csv')
```

```python
# df=df.iloc[:16,:17]
df=df.iloc[:21,:32]
df
```

### edit

```python
new_column_names = df.columns.tolist()
df.columns = range(len(df.columns))
```

```python
# Insert a new row at the beginning with the original column names
df.loc[-1] = new_column_names
```

```python
df.index = df.index + 1 
```

```python
df=df.sort_index().transpose()
df.columns = df.iloc[0]
df = df[1:]
df = df.rename(columns={'Cost Growth': 'Cost Digital', 'Cost Marketing ': 'Cost offline '})
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace(r'[+\-./]', '', regex=True)
df
```

```python
df.columns = ['report_day'] + df.columns.tolist()[1:]
df['report_day'] = pd.to_datetime(df['report_day'], format='%B %d')
df['report_day'] = df['report_day'].dt.strftime('2025-%m-%d')
df
```

```python
df['cac_incl_branding']=np.nan
df['fsfirst_disb']=np.nan
```

```python
df['first_disbursement_fidobiz'] = df['first_disb_fidobiz_migrated_new'].astype(float) + df['first_disbfidobiz_new'].astype(float)+df['first_disb_fidobiz_migrated_old'].astype(float)
```

```python
# List of columns to keep constant
constant_cols = ['report_day', 'new_installs', 'sign_up', 'kyc_completed', 'fs_eligible', 'first_disbursement_all',
                 'cost_digital', 'cost_offline_', 'cac', 'cac_incl_branding', 'cpi', 'cpl', 'cpkyc', 'installssignup',
                 'sign_upkyc_completed', 'kycfirst_disb', 'fsfirst_disb', 'new_install_first_dis',
                 'signupfirst_dis', 'first_disbursement_personal', 'first_disbursement_fidobiz',
                 'first_disbfidobiz_new', 'first_disb_fidobiz_migrated_old', 'first_disb_fidobiz_migrated_new',]

# Identify columns to move (those not in constant_cols)
cols_to_move = [col for col in df.columns if col not in constant_cols]

# Reorganize columns: keep constant columns first, then move the others
df = df[constant_cols + cols_to_move]
```
